```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Условие 1: Задача 1

### 1.1. Скачать данные по ссылке

```python
df = pd.read_csv('kc_final.csv')
df.head()
```

```python
df.describe()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(12,10))
sns.histplot(df['price'], bins = 15)
plt.title('Предложения объектов недвижимости разной стоимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество объектов')
plt.xlim(0,8000000)
```

### 1.3 Изучите распределение квадратуры жилой

```python
counts = df['sqft_living'].value_counts()
plt.scatter(counts.index, counts.values)
plt.title('Количество предложений разной жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Количество');
```

### 1.4 Изучите распределение года постройки

```python
counts = df['yr_built'].value_counts()
plt.figure(figsize=(20,5))
plt.title('Количество объектов по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов')
sns.barplot(x=counts.index, y=counts.values)
plt.xticks(rotation = 90)
plt.show();
```

### Выводы: В таблице преобладают предложения в ценовом сегменте до 1 млн. (примерно 2/3 от общего числа) жилой площади до 4.000 кв.футов (явный перевес в пользу помещений около 2.000 кв.футов). Объектов, построенных в период 1900-1950 г.г. существенно меньше, чем 1950-2015 годов постройки. Причем максимальное число предложение 2003-2006 годов постройки и 2014го.

## Условие 3: Задача 3

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

### Есть практически линейная зависимость стоимости объекта от его площади за редким исключением. Чем больше площадь дома, тем выше его цена.

```python
df_cut = df[df['price'] <= 2000000]
sns.jointplot(y=df['yr_built'], x=df_cut['price'], kind='kde');
```

### Анализ зависимости стоимости домов от года постройки показал относительно слабую корреляцию двух признаков. В ценовом диапазоне до 1,5 млн. представлены дома постройки от 1900х годов до 2020х с заметным преобладанием предложений после 1940го года и удорожанием вариантов >2000 года постройки. Однако явной корреляции не наблюдается.

```python
plt.figure(figsize=(25,10))
sns.histplot(x=df['view'], y=df['price']);
```

### Нет явной зависимости стоимости предложений от вида: дешевые варианты действительно имеют самый невыгодный вид. Вместе с тем по мере улучшения вида стоимость предложений не растет столь явно.

```python
plt.figure(figsize=(25,10))
df_cut = df[df['price'] <=2000000]
sns.boxplot(x=df['yr_renovated'], y=df_cut['price'])
plt.xticks(rotation = 90);
```

### Дома после реновации ценятся больше, чем без нее. При этом наиболее дорогие предложения консолидируются в области объектов, реновация которых была произведена в конце 1990х - начале 2000-х.

```python
plt.bar(df['floors'], df['price']);
```
